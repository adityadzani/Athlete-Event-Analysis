#### Top 10 Participated Countries

```python
data['Team'].value_counts(ascending = False).head(10)
```

```python
top_countries = pd.DataFrame(data['Team'].value_counts(ascending = False).head(10))
top_countries
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=top_countries.index, y=top_countries['Team'], palette= 'Set2')
plt.title('Top 10 Participated Countries', size = 20)
plt.xlabel('Country')
plt.ylabel('Participants')
plt.show()
```

#### Gold Medals for Athletes over 60 years

```python
data.columns
```

```python
gold = data[data['Medal'] == 'Gold']
gold
```

```python
gold = gold[np.isfinite(gold['Age'])]
gold
```

```python
sporting_event = gold[gold['Age'] > 60]['Sport']
sporting_event
```

```python
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(x = sporting_event)
plt.title('Gold Medals for Athletes over 60 years')
plt.yticks(np.arange(0,5, step=1))
plt.show()
```

#### Top 5 gold medals countries

```python
gold = data[data['Medal'] == 'Gold']
topRegion = gold[['Region']].value_counts().reset_index(name= 'Medal').head()
topRegion
```

```python
sns.catplot(x= 'Region', y= 'Medal', data = topRegion, 
           height= 5, kind= 'bar', palette= 'rocket')
plt.title('Top 5 gold medals countries')
plt.show()
```

#### Countrywise Medals for the year 2016

```python
top10 = data[(data['Year'] == data['Year'].max()) & 
     (data['Medal'] == 'Gold')]['Team'].value_counts().head(10).reset_index()
top10.rename(columns= {'index': 'Region', 'Team': 'Medal'}, inplace= True)
```

```python
sns.barplot(x = top10['Medal'], y = top10['Region'], palette = 'Set2')
plt.title('Countrywise Medals for the year 2016')
plt.show()
```
